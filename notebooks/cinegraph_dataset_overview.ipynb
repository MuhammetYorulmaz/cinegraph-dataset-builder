````python
import warnings
from pathlib import Path
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.ticker as mticker
import seaborn as sns

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 120)
pd.set_option("display.max_colwidth", 80)

# ── Cinematic warm palette ────────────────────────────────────────────────────
# Inspired by film poster art: warm golds, dusty roses, slate blues, sage greens
PALETTE = {
    "gold":     "#C9A84C",   # warm amber / award gold
    "rose":     "#C97B84",   # dusty rose
    "slate":    "#6B8CAE",   # muted slate blue
    "sage":     "#7A9E7E",   # sage green
    "terracotta": "#C07A5A", # warm terracotta
    "lavender": "#9B89AC",   # muted lavender
    "sand":     "#C4A882",   # warm sand
    "steel":    "#7B9BAE",   # steel blue
    "cream":    "#E8DECE",   # off-white cream
    "charcoal": "#4A4A4A",   # soft charcoal (for text/lines)
}

SEQ_MOVIE = [
    "#C9A84C", "#C07A5A", "#C97B84", "#9B89AC",
    "#6B8CAE", "#7B9BAE", "#7A9E7E", "#C4A882",
]
SEQ_TV    = [
    "#7A9E7E", "#9B89AC", "#6B8CAE", "#7B9BAE",
    "#C4A882", "#C07A5A", "#C9A84C", "#C97B84",
]

def set_cinematic_style():
    """Apply a warm, cinematic Matplotlib style."""
    mpl.rcParams.update({
        "figure.facecolor":  "#FAF7F2",
        "axes.facecolor":    "#FAF7F2",
        "axes.edgecolor":    "#D4CCBC",
        "axes.linewidth":    0.8,
        "axes.grid":         True,
        "grid.color":        "#E8E2D8",
        "grid.linewidth":    0.6,
        "grid.alpha":        0.7,
        "xtick.color":       "#5A5040",
        "ytick.color":       "#5A5040",
        "xtick.labelsize":   9,
        "ytick.labelsize":   9,
        "axes.labelcolor":   "#3A3028",
        "axes.labelsize":    10,
        "axes.titlesize":    12,
        "axes.titlecolor":   "#2A201C",
        "axes.titleweight":  "bold",
        "axes.titlepad":     12,
        "text.color":        "#3A3028",
        "legend.framealpha": 0.85,
        "legend.edgecolor":  "#D4CCBC",
        "legend.fontsize":   9,
        "figure.dpi":        110,
        "savefig.dpi":       150,
        "font.family":       "DejaVu Sans",
    })

set_cinematic_style()

# ── Data loading ─────────────────────────────────────────────────────────────
# Kaggle: files live at /kaggle/input/<dataset-name>/
# Local : adjust DATA_DIR to your local path
KAGGLE_INPUT = Path("/kaggle/input")
possible_dirs = list(KAGGLE_INPUT.glob("*")) if KAGGLE_INPUT.exists() else []
if possible_dirs:
    DATA_DIR = possible_dirs[0]           # first dataset folder on Kaggle
else:
    DATA_DIR = Path("../data")            # local fallback

def load(name: str) -> pd.DataFrame:
    path = DATA_DIR / name
    if not path.exists():
        print(f"  ⚠ missing: {name}  — returning empty DataFrame")
        return pd.DataFrame()
    for enc in ("utf-8-sig", "utf-8", "latin-1"):
        try:
            df = pd.read_csv(path, encoding=enc, low_memory=False)
            print(f"  ✓  {name:<28} {len(df):>7,} rows × {df.shape[1]} cols")
            return df
        except UnicodeDecodeError:
            continue
    raise RuntimeError(f"Could not decode {name}")

print("Loading dataset files...")
movies        = load("movies.csv")
tv_shows      = load("tv_shows.csv")
people        = load("people.csv")
orphan_movies = load("orphan_movies.csv")
orphan_tv     = load("orphan_tv.csv")
movie_reviews = load("movie_reviews.csv")
tv_reviews    = load("tv_reviews.csv")
print("\nAll files loaded.")
````

## 1 · Corpus Snapshot

A single-glance summary of all seven tables, their sizes, and key integrity metrics.


````python
# ── Table summary ─────────────────────────────────────────────────────────────
tables = [
    ("movies",        movies,        "tmdb_id"),
    ("tv_shows",      tv_shows,      "tmdb_id"),
    ("people",        people,        "tmdb_id"),
    ("orphan_movies", orphan_movies, "tmdb_id"),
    ("orphan_tv",     orphan_tv,     "tmdb_id"),
    ("movie_reviews", movie_reviews, "review_id"),
    ("tv_reviews",    tv_reviews,    "review_id"),
]

summary_rows = []
for name, df, pk in tables:
    if df.empty:
        summary_rows.append({"Table": name, "Rows": 0, "Columns": 0,
                              "PK duplicates": "—", "Memory (MB)": 0})
        continue
    dups = int(df[pk].duplicated().sum()) if pk in df.columns else "—"
    mem  = round(df.memory_usage(deep=True).sum() / 1024**2, 1)
    summary_rows.append({
        "Table":         name,
        "Rows":          f"{len(df):,}",
        "Columns":       df.shape[1],
        "PK duplicates": f"✅ 0" if dups == 0 else f"⚠ {dups}",
        "Memory (MB)":   mem,
    })

display(pd.DataFrame(summary_rows).set_index("Table").style
        .set_caption("Dataset table overview")
        .set_table_styles([{
            "selector": "caption",
            "props": [("font-size", "14px"), ("font-weight", "bold"),
                      ("color", "#3A3028"), ("text-align", "left"),
                      ("padding-bottom", "8px")]
        }])
        .set_properties(**{"text-align": "center"})
        .highlight_between(subset=["Memory (MB)"], left=0, right=9999,
                           props="background-color: #F0EDE6;")
)
````

````python
set_cinematic_style()
fig, axes = plt.subplots(1, 2, figsize=(15, 4.8))
fig.patch.set_facecolor("#FAF7F2")

for ax, frame, label, color, accent in [
    (axes[0], movies,   "Movies",   PALETTE["gold"],  "#8B6914"),
    (axes[1], tv_shows, "TV Shows", PALETTE["slate"], "#3A5A7A"),
]:
    years = pd.to_numeric(frame["release_year"], errors="coerce").dropna().astype(int)
    years = years[(years >= 1920) & (years <= 2027)]
    n, bins, patches = ax.hist(years, bins=range(1920, 2028, 3),
                               color=color, edgecolor="#FAF7F2",
                               linewidth=0.5, alpha=0.88)
    for patch, left in zip(patches, bins):
        intensity = 0.6 + 0.4 * (left - 1920) / (2027 - 1920)
        r, g, b, _ = mpl.colors.to_rgba(color)
        patch.set_facecolor((r * intensity, g * intensity, b * intensity))

    ax.set_title(f"{label}  ·  Release Year Distribution", fontsize=12, pad=14)
    ax.set_xlabel("Year", labelpad=6)
    ax.set_ylabel("Number of titles", labelpad=6)
    ax.set_xlim(1920, 2027)

    median_yr = int(years.median())
    ax.axvline(median_yr, color=accent, linewidth=1.4, linestyle="--", alpha=0.7)
    ax.text(median_yr + 1, ax.get_ylim()[1] * 0.92,
            f"median\n{median_yr}", color=accent, fontsize=8.5, va="top")

    # ← moved to top-LEFT to avoid the tall 2020s bars on the right
    ax.text(0.03, 0.97, f"n = {len(years):,}",
            transform=ax.transAxes,
            ha="left", va="top", fontsize=9, color="#7A6840",
            bbox=dict(boxstyle="round,pad=0.35", fc="#EFE8D8",
                      ec="#D4C8A8", alpha=0.9))

plt.suptitle("CineGraph — Temporal Coverage", fontsize=14, fontweight="bold",
             color="#2A201C", y=1.01)
plt.tight_layout()
plt.show()

print(f"Movies  span: {int(pd.to_numeric(movies['release_year'], errors='coerce').min())} – "
      f"{int(pd.to_numeric(movies['release_year'], errors='coerce').max())}")
print(f"TV shows span: {int(pd.to_numeric(tv_shows['release_year'], errors='coerce').min())} – "
      f"{int(pd.to_numeric(tv_shows['release_year'], errors='coerce').max())}")
````

<a id="22-language-distribution"></a>
### 2.2 Language Distribution

````python
set_cinematic_style()
fig, axes = plt.subplots(1, 2, figsize=(15, 5.2))
fig.patch.set_facecolor("#FAF7F2")

LANG_LABELS = {
    "en": "English", "fr": "French", "es": "Spanish", "ja": "Japanese",
    "it": "Italian", "ko": "Korean", "de": "German", "zh": "Chinese",
    "ru": "Russian", "pt": "Portuguese", "cn": "Cantonese", "hi": "Hindi",
    "nl": "Dutch",   "th": "Thai",     "tr": "Turkish",  "ar": "Arabic",
}

for ax, frame, label, cmap_base, n_top in [
    (axes[0], movies,   "Movies",   PALETTE["gold"],  12),
    (axes[1], tv_shows, "TV Shows", PALETTE["slate"], 12),
]:
    top = frame["original_language"].value_counts().head(n_top)
    nice_labels = [LANG_LABELS.get(k, k.upper()) for k in top.index]
    total = len(frame)

    # Build a warm gradient of bars
    base = mpl.colors.to_rgb(cmap_base)
    colors_bar = [
        tuple(c * (0.55 + 0.45 * i / len(top)) for c in base)
        for i in range(len(top))
    ]

    bars = ax.barh(nice_labels, top.values, color=colors_bar,
                   edgecolor="#FAF7F2", linewidth=0.5, height=0.72)
    for bar, val in zip(bars, top.values):
        ax.text(bar.get_width() + total * 0.003, bar.get_y() + bar.get_height() / 2,
                f"{val / total * 100:.1f}%",
                va="center", fontsize=8.5, color="#5A4830")

    ax.set_title(f"{label}  ·  Top {n_top} Original Languages", fontsize=12, pad=14)
    ax.set_xlabel("Number of titles", labelpad=6)
    ax.set_xlim(0, top.values.max() * 1.18)
    ax.invert_yaxis()
    ax.tick_params(axis="y", labelsize=9.5)

plt.suptitle("CineGraph — Language Diversity  (70 + languages total)",
             fontsize=14, fontweight="bold", color="#2A201C", y=1.01)
plt.tight_layout()
plt.show()
````

<a id="23-genre-landscape"></a>
### 2.3 Genre Landscape

````python
def explode_genres(frame: pd.DataFrame) -> Counter:
    c = Counter()
    for v in frame["genres"].dropna():
        for g in str(v).split(","):
            g = g.strip()
            if g:
                c[g] += 1
    return c

set_cinematic_style()
fig, axes = plt.subplots(1, 2, figsize=(15, 6.2))
fig.patch.set_facecolor("#FAF7F2")

for ax, frame, label, palette_list in [
    (axes[0], movies,   "Movies",   SEQ_MOVIE),
    (axes[1], tv_shows, "TV Shows", SEQ_TV),
]:
    ctr  = explode_genres(frame)
    top  = pd.Series(dict(ctr.most_common(14))).sort_values()
    total = len(frame)

    # Assign one colour per genre, cycling through palette
    bar_colors = [palette_list[i % len(palette_list)] for i in range(len(top))]
    bars = ax.barh(top.index, top.values, color=bar_colors,
                   edgecolor="#FAF7F2", linewidth=0.5, height=0.75)

    for bar, val in zip(bars, top.values):
        ax.text(bar.get_width() + total * 0.004, bar.get_y() + bar.get_height() / 2,
                f"{val:,}  ({val / total * 100:.0f}%)",
                va="center", fontsize=8.5, color="#5A4830")

    ax.set_title(f"{label}  ·  Top 14 Genres", fontsize=12, pad=14)
    ax.set_xlabel("Title count  (titles may span multiple genres)", labelpad=6)
    ax.set_xlim(0, top.values.max() * 1.28)
    ax.tick_params(axis="y", labelsize=9.5)

plt.suptitle("CineGraph — Genre Landscape", fontsize=14, fontweight="bold",
             color="#2A201C", y=1.01)
plt.tight_layout()
plt.show()
````

### 3.1 Budget vs. Revenue Overview


````python
fin = movies[movies["budget_usd"].notna() & movies["revenue_usd"].notna()].copy()
fin = fin[fin["budget_usd"] > 0]
fin["roi_pct"]   = pd.to_numeric(fin["roi_pct"],   errors="coerce")
fin["budget_m"]  = fin["budget_usd"]  / 1e6
fin["revenue_m"] = fin["revenue_usd"] / 1e6
fin["profit_m"]  = fin["profit_usd"]  / 1e6

print(f"Films with full financials: {len(fin):,}  "
      f"({len(fin) / len(movies) * 100:.1f}% of corpus)")
print()
display(fin[["budget_usd", "revenue_usd", "profit_usd", "roi_pct"]]
        .describe().round(1)
        .rename(index={"count":"count","mean":"mean","std":"std",
                        "min":"min","25%":"Q1","50%":"median",
                        "75%":"Q3","max":"max"})
        .style.format("{:,.1f}")
        .background_gradient(cmap="YlOrBr", axis=0)
        .set_caption("Financial summary — films with known budget & revenue")
)
````

````python
def count_providers(frame, col, top_n=12):
    ctr = Counter()
    for v in frame.get(col, pd.Series()).dropna():
        for name in str(v).split(","):
            n = name.strip()
            if n:
                ctr[n] += 1
    return pd.Series(dict(ctr.most_common(top_n)))

set_cinematic_style()
fig, axes = plt.subplots(1, 2, figsize=(15, 5.5))
fig.patch.set_facecolor("#FAF7F2")

for ax, col, title, base_color in [
    (axes[0], "watch_tr_flatrate", "🇹🇷 Turkey  ·  Top Subscription Services", PALETTE["terracotta"]),
    (axes[1], "watch_us_flatrate", "🇺🇸 United States  ·  Top Subscription Services", PALETTE["slate"]),
]:
    top_m = count_providers(movies,   col, 10).sort_values()
    top_t = count_providers(tv_shows, col, 10).sort_values()

    # Merge by service name
    all_svc = sorted(set(top_m.index) | set(top_t.index), key=lambda s: top_m.get(s,0)+top_t.get(s,0))
    mv_vals = [top_m.get(s, 0) for s in all_svc]
    tv_vals = [top_t.get(s, 0) for s in all_svc]

    y    = np.arange(len(all_svc))
    w    = 0.38
    base = mpl.colors.to_rgb(base_color)
    lighter = tuple(min(1, c + 0.25) for c in base)

    ax.barh(y - w/2, mv_vals, w, color=base_color, edgecolor="#FAF7F2", label="Movies",   alpha=0.88)
    ax.barh(y + w/2, tv_vals, w, color=lighter,    edgecolor="#FAF7F2", label="TV Shows", alpha=0.88)
    ax.set_yticks(y)
    ax.set_yticklabels(all_svc, fontsize=9.5)
    ax.set_xlabel("Number of titles", labelpad=6)
    ax.set_title(title, fontsize=11, pad=12)
    ax.legend(fontsize=9)

plt.suptitle("CineGraph — Streaming Market Coverage",
             fontsize=14, fontweight="bold", color="#2A201C", y=1.01)
plt.tight_layout()
plt.show()
````

<a id="5-recommendation-graph-coverage"></a>
## 5 · Recommendation-Graph Coverage

TMDB stores two link types per title: `similar_ids` (editorial similarity) and `recommended_ids` (algorithmic).  
Both lists are preserved in CineGraph. The **orphan tables** exist precisely to close dangling edges.


````python
def parse_id_list(value):
    if pd.isna(value): return []
    parsed = []
    for tok in str(value).split(","):
        tok = tok.strip()
        if not tok: continue
        try:    parsed.append(int(float(tok)))
        except: pass
    return parsed

def coverage_pct(frame, col, universe):
    total = hit = 0
    for v in frame.get(col, pd.Series()):
        ids  = parse_id_list(v)
        total += len(ids)
        hit   += sum(1 for i in ids if i in universe)
    return hit / total * 100 if total else 0

movie_main = set(movies["tmdb_id"].dropna().astype(int))
movie_full = movie_main | (set(orphan_movies["tmdb_id"].dropna().astype(int))
                            if not orphan_movies.empty else set())
tv_main    = set(tv_shows["tmdb_id"].dropna().astype(int))
tv_full    = tv_main    | (set(orphan_tv["tmdb_id"].dropna().astype(int))
                            if not orphan_tv.empty else set())

cov_data = {
    "Reference":     ["movies · similar_ids",   "movies · recommended_ids",
                      "tv_shows · similar_ids", "tv_shows · recommended_ids"],
    "Main corpus":   [coverage_pct(movies,   "similar_ids",    movie_main),
                      coverage_pct(movies,   "recommended_ids",movie_main),
                      coverage_pct(tv_shows, "similar_ids",    tv_main),
                      coverage_pct(tv_shows, "recommended_ids",tv_main)],
    "With orphans":  [coverage_pct(movies,   "similar_ids",    movie_full),
                      coverage_pct(movies,   "recommended_ids",movie_full),
                      coverage_pct(tv_shows, "similar_ids",    tv_full),
                      coverage_pct(tv_shows, "recommended_ids",tv_full)],
}
cov_df = pd.DataFrame(cov_data).set_index("Reference")

set_cinematic_style()
fig, ax = plt.subplots(figsize=(12, 4.8))

x     = np.arange(len(cov_df))
width = 0.38
b1 = ax.bar(x - width/2, cov_df["Main corpus"],  width,
            color=PALETTE["sand"],  edgecolor="#FAF7F2", label="Main corpus only", alpha=0.88)
b2 = ax.bar(x + width/2, cov_df["With orphans"], width,
            color=PALETTE["sage"],  edgecolor="#FAF7F2", label="Main + orphan tables", alpha=0.88)

for bars in (b1, b2):
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 1,
                f"{h:.1f}%", ha="center", va="bottom", fontsize=9, fontweight="bold",
                color="#3A3028")

ax.set_xticks(x)
ax.set_xticklabels(cov_df.index, rotation=10, ha="right", fontsize=10)
ax.set_ylim(0, 110)
ax.set_ylabel("Graph edges that resolve (%)", labelpad=8)
ax.set_title("Recommendation-Graph Resolution Rate\nOrphan tables close nearly all dangling edges for recommended_ids",
             fontsize=12, pad=14)
ax.axhline(100, color=PALETTE["charcoal"], linewidth=0.7, linestyle=":", alpha=0.4)
ax.legend(fontsize=9.5)

plt.tight_layout()
plt.show()

print("\nNumerical summary:")
display(cov_df.round(1).style
        .background_gradient(cmap="YlOrBr", axis=None)
        .format("{:.1f}%"))
````

<a id="6-cast-people-join-integrity"></a>
## 6 · Cast & People Join Integrity

People are collected *by reference* — only cast/directors who appear in the main corpus are stored.  
The coverage percentages reflect how many unique IDs from the title tables resolve in `people.csv`.

<a id="61-join-coverage"></a>
### 6.1 Join Coverage


````python
people_ids = set(people["tmdb_id"].dropna().astype(int))

def join_stats(frame, col):
    all_ids = set()
    for v in frame.get(col, pd.Series()):
        all_ids.update(parse_id_list(v))
    if not all_ids:
        return 0, 0, 0.0
    hit = len(all_ids & people_ids)
    return hit, len(all_ids), hit / len(all_ids) * 100

join_rows = [
    ("movies · cast_ids",      *join_stats(movies,   "cast_ids")),
    ("movies · director_ids",  *join_stats(movies,   "director_ids")),
    ("tv_shows · cast_ids",    *join_stats(tv_shows, "cast_ids")),
    ("tv_shows · creator_ids", *join_stats(tv_shows, "creator_ids")),
]
join_df = pd.DataFrame(join_rows,
                        columns=["Relation", "Resolved", "Total unique", "Coverage %"])
join_df["Coverage %"] = join_df["Coverage %"].round(1)

set_cinematic_style()
fig, ax = plt.subplots(figsize=(13, 5))

x = np.arange(len(join_df))
w = 0.38
ax.bar(x - w/2, join_df["Total unique"], w,
       color=PALETTE["sand"],  edgecolor="#FAF7F2", label="Unique references", alpha=0.88)
ax.bar(x + w/2, join_df["Resolved"],    w,
       color=PALETTE["sage"],  edgecolor="#FAF7F2", label="Resolved in people.csv", alpha=0.88)

ax.set_xticks(x)
ax.set_xticklabels(join_df["Relation"], rotation=10, ha="right", fontsize=10.5)
ax.set_ylabel("Unique person IDs", labelpad=8)
ax.set_title("Cast & Creator References  vs.  Entries Resolved in people.csv",
             fontsize=12, pad=14)
ax.legend(fontsize=9.5)

for i, row in join_df.iterrows():
    ax.text(i, max(join_df["Total unique"]) * 1.02,
            f'{row["Coverage %"]:.1f}%',
            ha="center", fontsize=11, fontweight="bold", color=PALETTE["sage"])

plt.tight_layout()
plt.show()

display(join_df.style
        .format({"Resolved": "{:,}", "Total unique": "{:,}", "Coverage %": "{:.1f}%"})
        .background_gradient(subset=["Coverage %"], cmap="YlOrBr")
        .set_caption("People join coverage by reference column"))
````

````python
movie_refs = set()
for col in ["cast_ids", "director_ids"]:
    if col in movies.columns:
        for v in movies[col]: movie_refs.update(parse_id_list(v))

tv_refs = set()
for col in ["cast_ids", "creator_ids"]:
    if col in tv_shows.columns:
        for v in tv_shows[col]: tv_refs.update(parse_id_list(v))

ppl = set(people["tmdb_id"].dropna().astype(int))
counts = {
    "Movies only":    len(ppl & movie_refs - tv_refs),
    "TV only":        len(ppl & tv_refs - movie_refs),
    "Both":           len(ppl & movie_refs & tv_refs),
}

set_cinematic_style()
fig, ax = plt.subplots(figsize=(7, 7))
wedge_colors = [PALETTE["gold"], PALETTE["slate"], PALETTE["sage"]]
wedges, texts, autotexts = ax.pie(
    counts.values(),
    labels=counts.keys(),
    colors=wedge_colors,
    autopct="%1.1f%%",
    pctdistance=0.78,
    startangle=120,
    wedgeprops={"edgecolor": "#FAF7F2", "linewidth": 2},
)
for t in autotexts:
    t.set_fontsize(11); t.set_fontweight("bold"); t.set_color("#2A201C")
for t in texts:
    t.set_fontsize(11); t.set_color("#3A3028")

centre = plt.Circle((0, 0), 0.52, fc="#FAF7F2")
ax.add_patch(centre)
ax.text(0, 0.06, f"{len(ppl):,}", ha="center", va="center",
        fontsize=22, fontweight="bold", color="#2A201C")
ax.text(0, -0.16, "people total", ha="center", va="center",
        fontsize=10, color="#7A6840")

ax.set_title("People.csv — Corpus Span per Person", fontsize=12, pad=20)
plt.tight_layout()
plt.show()

for k, v in counts.items():
    print(f"  {k:<18}: {v:>6,}  ({v/len(ppl)*100:.1f}%)")
````

## 9 · Data-Quality Audit

Null rates across the columns that matter most for downstream tasks.  
Raw-delivery philosophy: **nulls are preserved as-is** so analysts can apply their own imputation strategy.


````python
critical = {
    "movies":   ["title", "overview", "genres", "cast_names", "directors",
                 "vote_average", "release_year", "budget_usd", "us_certification"],
    "tv_shows": ["title", "overview", "genres", "cast_names", "creators",
                 "vote_average", "release_year", "cert_us"],
    "people":   ["name", "biography", "known_for_dept", "birthday",
                 "known_movies", "profile_url"],
}

audit_rows = []
for table, cols in critical.items():
    frame = {"movies": movies, "tv_shows": tv_shows, "people": people}[table]
    for col in cols:
        if col not in frame.columns: continue
        pct = frame[col].replace(r"^\s*$", np.nan, regex=True).isna().mean() * 100
        audit_rows.append({"Table": table, "Column": col, "Null %": round(pct, 1)})

audit_df = pd.DataFrame(audit_rows)
pivot_q = (audit_df.pivot(index="Column", columns="Table", values="Null %")
           .reindex(columns=["movies", "tv_shows", "people"]))

set_cinematic_style()
fig, ax = plt.subplots(figsize=(11, 7))

from matplotlib.colors import LinearSegmentedColormap
cmap_audit = LinearSegmentedColormap.from_list(
    "audit", ["#EFE8D8", PALETTE["gold"], PALETTE["terracotta"]], N=256
)

im = ax.imshow(pivot_q.fillna(-1).values, cmap=cmap_audit,
               aspect="auto", vmin=0, vmax=60)

# Annotate — text-only, no emoji
for i in range(pivot_q.shape[0]):
    for j in range(pivot_q.shape[1]):
        val = pivot_q.iloc[i, j]
        if np.isnan(val):
            ax.text(j, i, "—", ha="center", va="center",
                    fontsize=11, color="#BBBBBB")
        else:
            # Colour-coded indicator: ● green / yellow / red
            if val <= 5:
                dot_color, dot = "#5A9E6A", "●"
            elif val <= 25:
                dot_color, dot = "#C9A84C", "●"
            else:
                dot_color, dot = "#C07A5A", "●"
            text_color = "#2A201C" if val < 30 else "#FAF7F2"
            # Value on first line
            ax.text(j, i - 0.14, f"{val:.1f}%",
                    ha="center", va="center", fontsize=9.5, color=text_color)
            # Coloured dot on second line
            ax.text(j, i + 0.22, dot,
                    ha="center", va="center", fontsize=9, color=dot_color)

ax.set_xticks(range(3))
ax.set_xticklabels(["Movies", "TV Shows", "People"], fontsize=11, fontweight="bold")
ax.set_yticks(range(len(pivot_q.index)))
ax.set_yticklabels(pivot_q.index, fontsize=10)

# Legend as text inside title area — no emoji
ax.set_title(
    "Null Rate Heatmap — Critical Columns\n"
    "● green ≤5%   ● amber 5–25%   ● red >25%",
    fontsize=12, pad=16
)

import matplotlib.patches as mpatches
leg_handles = [
    mpatches.Patch(color="#5A9E6A", label="≤ 5%  (low / expected)"),
    mpatches.Patch(color="#C9A84C", label="5–25%  (moderate)"),
    mpatches.Patch(color="#C07A5A", label="> 25%  (high / structural)"),
]
ax.legend(handles=leg_handles, loc="lower right", fontsize=9,
          framealpha=0.9, edgecolor="#D4CCBC")

ax.set_title("Null Rate Heatmap — Critical Columns", fontsize=12, pad=16)

cbar = plt.colorbar(im, ax=ax, fraction=0.025, pad=0.03)
cbar.set_label("Null %", fontsize=10)
cbar.ax.tick_params(labelsize=9)

plt.tight_layout()
plt.show()

print("\n Notable patterns:")
print("  • tv_shows.creators  ~39 %  null — many series list no credited creator on TMDB (expected)")
print("  • people.biography   ~39 %  null — especially common for minor/crew roles (expected)")
print("  • movies.budget_usd  ~57 %  null — TMDB does not publish unreported budgets (expected)")
print("  • All primary keys   100 %  non-null with zero duplicates  [OK]")
````

### 11.2 Parsing Comma-Separated ID Columns

````python
def parse_id_list(value):
    """Turn '123, 456, 789' → [123, 456, 789]."""
    if pd.isna(value): return []
    parsed = []
    for tok in str(value).split(","):
        tok = tok.strip()
        if not tok: continue
        try:    parsed.append(int(float(tok)))
        except: pass
    return parsed

# Example — explode cast for network analysis
cast_edges = (
    movies[["tmdb_id", "title", "cast_ids"]].dropna(subset=["cast_ids"])
    .assign(cast_id=lambda df: df["cast_ids"].str.split(","))
    .explode("cast_id")
    .assign(cast_id=lambda df: pd.to_numeric(df["cast_id"].str.strip(), errors="coerce"))
    .dropna(subset=["cast_id"])
    .astype({"cast_id": int})
)
print(f"Cast edge rows (movie × person pairs): {len(cast_edges):,}")
print(cast_edges[["title", "cast_id"]].head(8).to_string(index=False))
````

### 11.3 Joining Titles with People

````python
people_idx = people.set_index("tmdb_id")[["name", "known_for_dept", "profile_url"]]

cast_joined = cast_edges.merge(people_idx, left_on="cast_id", right_index=True, how="left")
resolved = cast_joined["name"].notna().sum()
print(f"Resolved: {resolved:,} / {len(cast_joined):,}  ({resolved/len(cast_joined)*100:.1f}%)")
print()
print(cast_joined[cast_joined["name"].notna()][["title","name","known_for_dept"]].head(12)
      .to_string(index=False))
````

<a id="important-data-notes"></a>
### Important data notes

- **List columns** (`cast_ids`, `genres`, `keywords`, …) are stored as **comma-separated strings**. Use `parse_id_list()` (Section 11.2) to split them.
- **Financial columns** (`budget_usd`, `roi_pct`, …) are `null` when TMDB does not report both budget and revenue — this is normal, not a pipeline error.
- **`creators`** in `tv_shows.csv` is null for ~39 % of series — this reflects TMDB's own database, not a collection failure.
- **Primary keys** (`tmdb_id` everywhere, `review_id` in reviews) are **100 % non-null** with **zero duplicates** across all tables.
- All files are **UTF-8 with BOM** (`encoding="utf-8-sig"`) for seamless Excel compatibility.
